```python
import sympy
from tensorconvertor import FourthOrderTensor

sympy.init_printing(use_latex=True)
```

## 3-d

```python
def rotation(x):
    r = sympy.MatrixSymbol("R", x.shape[0], x.shape[0])
    return r.T * x * r


FourthOrderTensor().from_operator(rotation).as_mandel()
```

```python
FourthOrderTensor().from_operator(rotation).as_unsym()
```

## 2-d

```python
def rotation(x):
    theta = sympy.symbols("theta")
    sin = sympy.sin(theta)
    cos = sympy.cos(theta)
    r = sympy.Matrix([[cos, -sin], [sin, cos]])
    return r.T * x * r


sympy.expand_trig(FourthOrderTensor(dim=2).from_operator(rotation).as_mandel())
```

```python
sympy.expand_trig(FourthOrderTensor(dim=2).from_operator(rotation).as_unsym())
```
